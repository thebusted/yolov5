```python
import glob
import shutil
import os
from pathlib import Path

# Get a list of all image files in the directory and its subdirectories
image_files = glob.glob('E:\dataset\cattle\BeefCattle_Muzzle\*\*', recursive=True)

# Output the list of image files
dataset_images_dir = 'E:\dataset\cattle\BeefCattle_Muzzle_dataset\images'
dataset_labels_dir = 'E:\dataset\cattle\BeefCattle_Muzzle_dataset\labels'

# Print the list of image files
for image_file in image_files:
    print(image_file)
    
    # Copy image to dataset_images_dir
    shutil.copy(image_file, dataset_images_dir)
    
    # Create a .txt file for each image file in dataset_labels_dir
    base_name = Path(image_file).stem
    txt_file_path = os.path.join(dataset_labels_dir, f"{base_name}.txt")
    with open(txt_file_path, 'w') as f:
        f.write("0 0.5 0.5 1.0 1.0\n")  # Write the data to the .txt file
    
print("Total images", len(image_files), "files")
```
